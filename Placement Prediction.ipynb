# Data Preprocessing

```python
data = data.drop(['sl_no', 'salary'], axis=1)
```

```python
data
```

```python
data['ssc_b'].unique()
```

```python
data['ssc_b']=data['ssc_b'].map({'Central':1,'Others':0})
```

```python
data['hsc_b'].unique()
```

```python
data['hsc_b']=data['hsc_b'].map({'Central':1,'Others':0})
```

```python
data['hsc_s'].unique()
```

```python
data['hsc_s']=data['hsc_s'].map({'Commerce':1, 'Science':2, 'Arts':0})
```

```python
data['degree_t'].unique()
```

```python
data['degree_t']=data['degree_t'].map({'Sci&Tech':2, 'Comm&Mgmt':1, 'Others':0})
```

```python
data['specialisation'].unique()
```

```python
data['specialisation']=data['specialisation'].map({'Mkt&HR':1, 'Mkt&Fin':0})
```

```python
data['workex'].unique()
```

```python
data['workex']=data['workex'].map({'No':0, 'Yes':1})
```

```python
data['status']=data['status'].map({'Placed':1, 'Not Placed':0})
```

```python
data
```

# storing featurs Matrix in X and Response(Target ) In y

```python
X=data.drop('status',axis=1)
y=data['status']
```

# Spliting data into  training sets and test sets

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.20,random_state=42)
```

# importing Ml model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn import svm
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

# Model training

```python
lr=LogisticRegression()
lr.fit(X_train,y_train)
svm=svm.SVC()
svm.fit(X_train,y_train)
knn=KNeighborsClassifier()
knn.fit(X_train,y_train)
dt=DecisionTreeClassifier()
dt.fit(X_train,y_train)
rf=RandomForestClassifier()
rf.fit(X_train,y_train)
gb=GradientBoostingClassifier()
gb.fit(X_train,y_train)
```

# Prediction on test data

```python
y_pred1=lr.predict(X_test)
y_pred2=svm.predict(X_test)
y_pred3=knn.predict(X_test)
y_pred4=dt.predict(X_test)
y_pred5=rf.predict(X_test)
y_pred6=gb.predict(X_test)
```

# evaluating the algo

```python
from sklearn.metrics import accuracy_score
```

```python
score1=accuracy_score(y_test,y_pred1)
score2=accuracy_score(y_test,y_pred2)
score3=accuracy_score(y_test,y_pred3)
score4=accuracy_score(y_test,y_pred4)
score5=accuracy_score(y_test,y_pred5)
score6=accuracy_score(y_test,y_pred6)
```

```python
print(score1,score2,score3,score4,score5,score6)
```

```python
final_data=pd.DataFrame({'Model':['LR','SVC','KNN','DT','RF','GB'],
                        'ACC':[score1*100,score2*100,score3*100,score4*100,score5*100,score6*100]})
```
